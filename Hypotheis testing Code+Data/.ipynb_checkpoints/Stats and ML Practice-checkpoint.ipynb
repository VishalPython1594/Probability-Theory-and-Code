```python
from scipy.stats import ttest_1samp
```

```python
import numpy as np
age = [20,25,30,31,23,45,32,43,32,23,24,19,28,24,30,30,23,22,20,20,20,20,21,20]
```

```python
mean_age = np.mean(age)
t,pv = ttest_1samp(age, 27)
print(t)
print(pv)
```

```python
if pv < 0.05:    # significance is 0.05
    print("Reject the null hypothesis")
else:
    print("Failed to reject the null hypothesis")
```

```python
from scipy.stats import ttest_ind
```

```python
data_1 = np.random.uniform(low = 25, high= 75, size = (20,))
data_2 = np.random.uniform(12, 52.5, size = (20,))

t, p = ttest_ind(data_1,data_2)
print('data_1', data_1)
print('data_2', data_2)
print(np.mean(data_1))
print(np.mean(data_2))
print()
print('t-statistic', t)
print('p-value', p)

if pv < 0.05:    # significance is 0.05
    print("Reject the null hypothesis")
else:
    print("Failed to reject the null hypothesis")
```

```python
y_sm.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sm, y_sm, test_size = 0.25)
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled_df = pd.DataFrame(x_train_scaled, columns = x.columns)
```

```python
x_testscaled_df = pd.DataFrame(x_test_scaled, columns = x.columns)
```

```python
nb = GaussianNB()
```

```python
model_nb = nb.fit(x_train_scaled_df, y_train)
```

```python
y_pred = nb.predict(x_testscaled_df)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
y_prob = model_nb.predict_proba(x_testscaled_df)[:,1]
```

```python
y_prob
```

```python
fpr, tpr, _ = roc_curve(y_test, y_prob)
```

```python
roc_auc = auc(fpr, tpr)
```

```python
roc_auc
```

```python
plt.plot(fpr,tpr,label = f'Area under curve is --> {round(roc_auc,2)}')
plt.plot(fpr,fpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Roc-Auc Curve')
plt.legend()
plt.show()
```
