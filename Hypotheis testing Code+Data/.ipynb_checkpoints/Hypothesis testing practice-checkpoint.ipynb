```python
from scipy.stats import ttest_ind
```

```python
data1 = np.random.uniform(low=20, high=25.3, size=(20,))
print(data1)
```

```python
data2 = np.random.uniform(low=20.5, high=26.5, size=(20,))

data1_avg = np.mean(data1)
data2_avg = np.mean(data2)

print(data1_avg)
print(data2_avg)

data1_sd = np.std(data1)
data2_sd = np.std(data2)

print(data1_sd)
print(data2_sd)
```

```python
tt,pva = ttest_ind(data1,data2)
print(pva)
if pva<0.05:
    print('reject the null hypothesis')
else:
    print('failed to reject null hypothesis')
    
```

```python
import numpy as np
import pandas as pd

#create data
df = pd.DataFrame({'water': np.repeat(['daily', 'weekly'], 15),
                   'sun': np.tile(np.repeat(['low', 'med', 'high'], 5), 2),
                   'height': [6, 6, 6, 5, 6, 5, 5, 6, 4, 5,
                              6, 6, 7, 8, 7, 3, 4, 4, 4, 5,
                              4, 4, 4, 4, 4, 5, 6, 6, 7, 8]})
```

```python
df
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#perform two-way ANOVA
model = ols('height ~ C(water) + C(sun) + C(water) and C(sun)', data=df).fit()
res = sm.stats.anova_lm(model, typ=2)
res
```

```python
res['PR(>F)']
```

## Assignment solution

```python
for i in res['PR(>F)']:
    if isinstance(i,float) and i < 0.05:
        print('reject the null hypothesis')
    else:
        print('failed to reject the null hypothesis')
```

```python
df = pd.DataFrame({'Pateint':[1,2,3,4,5],'Sex':['Male','Male','Male','Male','Male'],'age-grp':['30-45','30-45','30-45','30-45','30-45'],
                   'bp_before':[143,163,153,153,146],'bp_after':[153,170,168,142,141]})
df
```

```python
from scipy import stats
```

```python
ttest,pval = stats.ttest_rel(df['bp_before'],df['bp_after'])
print(pval)
```

```python
if pval<0.05:
    print("reject null hypothesis")
else:
    print("failed to reject the null hypothesis")
```

## One Sample Z Test

```python
from statsmodels.stats import weightstats as stests
```

```python
ztest,pval = stests.ztest(df['bp_before'],x2 = None, value=152)
print(pval)
```

## 2 Sample Z Test

```python
ztest ,pval1 = stests.ztest(df['bp_before'], x2=df['bp_after'], value=0,alternative='two-sided')
print(pval1)
if pval1<0.05:
    print("reject null hypothesis")
else:
    print("failed to reject the null hypothesis")
```

Here is an example on a shell measurement (the length of the anterior adductor muscle scar, standardized by dividing by length) in the mussel Mytilus trossulus from five locations: Tillamook, Oregon; Newport, Oregon; Petersburg, Alaska; Magadan, Russia; and Tvarminne, Finland, taken from a much larger data set used in McDonald et al. (1991)

### One Way Anova

```python
from scipy.stats import f_oneway
```

```python
tillamook = [0.0571, 0.0813, 0.0831, 0.0976, 0.0817, 0.0859, 0.0735,
             0.0659, 0.0923, 0.0836]
newport = [0.0873, 0.0662, 0.0672, 0.0819, 0.0749, 0.0649, 0.0835,
           0.0725]
petersburg = [0.0974, 0.1352, 0.0817, 0.1016, 0.0968, 0.1064, 0.105]
magadan = [0.1033, 0.0915, 0.0781, 0.0685, 0.0677, 0.0697, 0.0764,
           0.0689]
tvarminne = [0.0703, 0.1026, 0.0956, 0.0973, 0.1039, 0.1045]

F,p = f_oneway(tillamook, newport, petersburg, magadan, tvarminne)
print("P Value calculated is ", p)

if p<0.05:
    print("We will reject the null hypothesis")
else:
    print("failed to reject the null hypothesis")
```
